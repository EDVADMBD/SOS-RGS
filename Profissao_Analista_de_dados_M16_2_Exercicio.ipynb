## 4\. Manipulação

```
# faça o código de manipulação de dados:
#
# - enriquecimento;
# - controle de qualidade;
# - etc.

# Adicionar uma coluna para representar a média de itens recebidos por habitante
populacao_habitantes = [1484941, 39992, 347613, 81812, 117480, 62771, 19833, 32273, 85371, 282558]  # População hipotética por cidade
df_doacoes_rs['Itens_por_habitante'] = df_doacoes_rs['Itens_recebidos'] / populacao_habitantes

# Verificar e lidar com valores ausentes
valores_ausentes = df_doacoes_rs.isnull().sum()
if valores_ausentes.any():
    print("Valores ausentes encontrados. Realizando tratamento...")
    # Para este exemplo, vamos preencher os valores ausentes com a média dos valores existentes
    df_doacoes_rs.fillna(df_doacoes_rs.mean(), inplace=True)
    print("Tratamento concluído. Valores ausentes preenchidos com a média.")
else:
    print("Não foram encontrados valores ausentes no DataFrame.")
```

```
#realizar a manipulação de dados e calcular as estatísticas
# Calcular estatísticas descritivas para a coluna "Itens_recebidos"
estatisticas_descritivas = df_doacoes_rs["Itens_recebidos"].describe()

# Exibir as estatísticas descritivas
print("Estatísticas Descritivas para Itens Recebidos:")
print(estatisticas_descritivas)

# Podemos realizar outras manipulações com base nessas estatísticas, por exemplo, filtrar as cidades com mais de 200.000 itens recebidos
cidades_mais_de_200k_itens = df_doacoes_rs[df_doacoes_rs["Itens_recebidos"] > 200000]
print("\nCidades com mais de 200.000 itens recebidos:")
print(cidades_mais_de_200k_itens)

# Ou ordenar o DataFrame pela quantidade de itens recebidos
df_doacoes_rs_ordenado = df_doacoes_rs.sort_values(by='Itens_recebidos', ascending=False)
print("\nDataFrame Ordenado pela quantidade de itens recebidos:")
print(df_doacoes_rs_ordenado)
```

```
# faça o código de visualização de dados:
#
# - produza pelo menos duas visualizações;
# - adicione um pequeno texto com os insights encontrados;
# - etc.

import seaborn as sns

# Criar o DataFrame com os dados fornecidos
dados = {
    'Cidade': ['Porto Alegre', 'Eldorado do Sul', 'Canoas', 'Lajeado', 'Guaíba', 'Montenegro', 'Cruzeiro do Sul', 'Estrela', 'Esteio', 'Santa Maria'],
    'Itens_recebidos': [792579, 336051, 282373, 189054, 121189, 97903, 72150, 64150, 63146, 57926]
}

df = pd.DataFrame(dados)

# Plotar o gráfico de variância
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Cidade', y='Itens_recebidos', palette='viridis')
plt.title('Itens Recebidos por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Itens Recebidos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# manipular esses dados em Python
import pandas as pd

# Ler o arquivo CSV
df = pd.read_csv('itens.csv', delimiter=';', header=None, names=['Item', 'Quantidade'])

# Mostrar os dados lidos
print("Dados Originais:")
print(df)

```

```
#visualizar os dados após processá-los
import matplotlib.pyplot as plt

# Gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df['Item'], df['Quantidade'], color='skyblue')
plt.xlabel('Item')
plt.ylabel('Quantidade')
plt.title('Quantidade de Itens')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# Ordenar o DataFrame por quantidade para uma melhor visualização da área
df_sorted = df.sort_values(by='Quantidade', ascending=False)

# Gráfico de área
plt.figure(figsize=(10, 6))
plt.fill_between(df_sorted['Item'], df_sorted['Quantidade'], color='skyblue', alpha=0.4)
plt.plot(df_sorted['Item'], df_sorted['Quantidade'], color='Slateblue', alpha=0.6)
plt.xlabel('Item')
plt.ylabel('Quantidade')
plt.title('Quantidade de Itens por Área')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

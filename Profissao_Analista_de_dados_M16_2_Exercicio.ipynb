## 3\. Exploração de dados

```
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Define o caminho do arquivo
file_path = 'doacoes_rs.csv'

# Verifica se o arquivo existe
if os.path.isfile(file_path):
    try:
        # Carregar o conjunto de dados
        df_doacoes_rs = pd.read_csv(file_path)
        print("Arquivo carregado com sucesso.")

        # Visualizar as primeiras linhas do DataFrame
        print("Visualização das primeiras linhas do DataFrame:")
        print(df_doacoes_rs.head())

        # Verificar informações sobre o DataFrame
        print("\nInformações sobre o DataFrame:")
        print(df_doacoes_rs.info())

        # Estatísticas descritivas dos dados numéricos
        print("\nEstatísticas descritivas dos dados numéricos:")
        print(df_doacoes_rs.describe())

        # Verificar se há valores ausentes
        print("\nValores ausentes no DataFrame:")
        print(df_doacoes_rs.isnull().sum())

        # Total de doações por cidade
        print("\nTotal de doações por cidade:")
        doacoes_por_cidade = df_doacoes_rs.groupby('cidade')['quantidade_doacoes'].sum().reset_index()
        print(doacoes_por_cidade.sort_values(by='quantidade_doacoes', ascending=False).head(10))

        # Visualização das 10 cidades com mais doações
        top_10_cidades = doacoes_por_cidade.sort_values(by='quantidade_doacoes', ascending=False).head(10)

        plt.figure(figsize=(12, 8))
        sns.barplot(x='quantidade_doacoes', y='cidade', data=top_10_cidades, palette='viridis')
        plt.title('Top 10 Cidades com Mais Doações')
        plt.xlabel('Quantidade de Doações')
        plt.ylabel('Cidade')
        plt.show()

    except Exception as e:
        print(f"Erro ao processar o arquivo: {e}")

else:
    print(f"Arquivo não encontrado em: {file_path}. Verifique o caminho e tente novamente.")



```

## 4\. Manipulação

```
# faça o código de manipulação de dados:
#
# - enriquecimento;
# - controle de qualidade;
# - etc.

# Adicionar uma coluna para representar a média de itens recebidos por habitante
populacao_habitantes = [1484941, 39992, 347613, 81812, 117480, 62771, 19833, 32273, 85371, 282558]  # População hipotética por cidade
df_doacoes_rs['Itens_por_habitante'] = df_doacoes_rs['Itens_recebidos'] / populacao_habitantes

# Verificar e lidar com valores ausentes
valores_ausentes = df_doacoes_rs.isnull().sum()
if valores_ausentes.any():
    print("Valores ausentes encontrados. Realizando tratamento...")
    # Para este exemplo, vamos preencher os valores ausentes com a média dos valores existentes
    df_doacoes_rs.fillna(df_doacoes_rs.mean(), inplace=True)
    print("Tratamento concluído. Valores ausentes preenchidos com a média.")
else:
    print("Não foram encontrados valores ausentes no DataFrame.")
```
